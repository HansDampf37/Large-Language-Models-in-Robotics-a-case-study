```python
# Visualize the results of:
#
# Zhao Mandi, Shreeya Jain, and Shuran Song. Roco: Dialectic multi-robot
# collaboration with large language models. In 2024 IEEE International
# Conference on Robotics and Automation (ICRA), pages 286â€“299. IEEE,
# 2024.

import matplotlib.pyplot as plt
import numpy as np
import os

# Tasks
tasks = [
    "Pack Grocery",
    "Arrange Cabinet",
    "Sweep Floor",
    "Make Sandwich",
    "Sort Cubes",
    "Move Rope"
]

# Methods and their success ratios
methods = [
    "Central Plan",
    "Dialog without History",
    "Dialog without Feedback",
    "Dialog"
]

success_ratios = {
    "Central Plan": [0.82, 0.90, 1.00, 0.96, 0.70, 0.50],
    "Dialog without History": [0.48, 1.00, 0.00, 0.33, 0.73, 0.65],
    "Dialog without Feedback": [0.35, 0.70, 0.95, 0.35, 0.53, 0.45],
    "Dialog": [0.44, 0.75, 0.95, 0.80, 0.93, 0.65],
}

# Plotting
x = np.arange(len(tasks))  # Task indices
width = 0.2  # Bar width

fig, ax = plt.subplots(figsize=(15, 5))

for i, method in enumerate(methods):
    ax.bar(
        x + i * width,
        success_ratios[method],
        width,
        label=method,
        alpha=0.8
    )

# Formatting
font_title = {'fontsize': 18}
font_labels = {'fontsize': 18}

# Adjusted x-ticks to be centralized under the grouped bars
ax.set_xticks(x + width * (len(methods) - 1) / 2)
ax.set_xticklabels(tasks, ha="center", fontsize=12)

ax.set_ylabel("Success Ratio", **font_labels)
ax.set_title("Success Ratios of Different Methods for Each Task", **font_title)
ax.legend(loc="best")
ax.grid(axis="y", linestyle="--", alpha=0.7)

# Ensure the 'images' directory exists
os.makedirs("images", exist_ok=True)

# Save the plot as a PNG file in the 'images' directory before displaying it
fig.savefig("images/roco_experiments.png")

# Show plot
plt.tight_layout()
plt.show()
```
