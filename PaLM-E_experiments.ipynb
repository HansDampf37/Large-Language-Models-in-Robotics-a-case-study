```python
# Visualize the results of:
#
# Driess, Danny, et al. "Palm-e: An embodied multimodal language model." arXiv preprint arXiv:2303.03378 (2023).

import matplotlib.pyplot as plt
import numpy as np
import os

# Training strategies
data_sets = [
    "Full mixture data",
    "Task specific data",
    "Full mixture data (LLM frozen)",
    "Task specific data (LLM frozen)",
]

data_set_results = [
    94.9,
    48.6,
    74.3,
    31.8
]
plt.figure(figsize=(15, 5))

# Formatting
font_title = {'fontsize': 20}
font_labels = {'fontsize': 13}
font_percentages = {'fontsize': 22}

# Plot the bars
bars = plt.bar(data_sets, data_set_results)

# Add numbers on the bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2.,  # Center the text horizontally
        height - 10,  # Position slightly below the top of the bar
        f'{height}%',  # Format as percentage
        ha='center', va='bottom', color='white', **font_percentages
    )

# Add labels and title
plt.title('Planning success results in the TAMP environment', **font_title)
plt.ylabel('Accuracy (%)', **font_labels)
plt.xticks(**font_labels)
plt.ylim(0, 100)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Ensure the 'images' directory exists
os.makedirs("images", exist_ok=True)

# Save the plot as a PNG file in the 'images' directory before displaying it
plt.savefig("images/palm-e_experiments.png")

# Display the plot
plt.show()
```
