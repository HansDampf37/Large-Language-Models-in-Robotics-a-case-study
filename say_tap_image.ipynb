```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load images
img1 = mpimg.imread('./input_images/foot_contact_patterns.png')
img2 = mpimg.imread('./input_images/example_prompts.png')

# Create subplots with different width ratios
fig, axes = plt.subplots(1, 2, figsize=(10, 5), gridspec_kw={'width_ratios': [3, 1]})

# Display images
axes[0].imshow(img1)
axes[0].axis('off')  # Hide axes for the first image
axes[0].set_title("Foot Contact Patterns")  # Set title for first image

axes[1].imshow(img2)
axes[1].axis('off')  # Hide axes for the second image
axes[1].set_title("Example Prompts")  # Set title for second image

# Adjust layout and show plot
plt.tight_layout()
# Ensure the 'images' directory exists
os.makedirs("images", exist_ok=True)

# Save the plot as a PNG file in the 'images' directory before displaying it
fig.savefig("images/say_tap_image.png")
plt.show()
```
